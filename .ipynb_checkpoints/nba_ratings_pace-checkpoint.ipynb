## 1. Terminology

### Possessions and Pace

According to Kubatko et al. in [SPABS](https://statsbylopez.files.wordpress.com/2016/01/jqas-2007-3-3-1070.pdf), possessions represent one of the most important concepts in basketball analysis. By their definition, a _possession_ is a stretch of continuous control of the ball by one team. Notably, an offensive rebound extends a possession, rather than create a new one.

While there are a few nuances that make defining posssessions non-trivial (such as free throws that terminate possesions), a simplification made in both BDS and SPABS gives:

$POSS = FGA + 0.44 \times FTA - OREB + TOV$

The 0.44 number is an estimate of the fraction of free throws that end possessions from the 2002-03 through 2005-06 NBA seasons. SPABS also discusses the concept of *plays*, which is different from possesions only in that plays do not include the *OREB* term.

With possessions defined, pace is simply the number of possessions per unit time. BDS uses *total* possessions per minute across all players on a team. [NBA.com](https://stats.nba.com/help/glossary/#pace) gives the usual definition of pace, which is the number of possessions per 48 minutes for a player or team.

### Ratings

Using possessions as the basis for assessing performance, offensive and defensive ratings for team _T_ against opponent _O_ follow as the number of points scored and surrendered per possession.

$ORtg_T = \frac{PTS_T}{POSS_T}$,
$DRtg_T = \frac{PTS_O}{POSS_O}$

### Data

The [stats.nba.com](https://stats.nba.com) API contains the `teamdashboardbygeneralsplits` endpoint that has tons of team performance data by different splits. The following leverages the `Team` class from the `py_ball` package to use data from the `teamdashboardbygeneralsplits` endpoint to replicate the pace and ratings visualizations in BDS. The `construct_season_logs` function abstracts the use of the `py_ball` to pull season data for a given team.

## 2. Fuctions and Analyses

```python
def construct_season_logs(team_id, measure_type='Base', league_id='00'):
    """ This function pulls 2020/2021 season data

    @param team_id (str): String of a 10-digit unique team identifier
    @param measure_type (str): One of 'Base', 'Advanced',
        'Misc', 'Scoring', 'Opponent', 'Usage' which dictates
        the type of statistics returned. 
    @param league_id (str): '00' for the NBA, '10' for the WNBA, 
        '20' for the G-League

    Returns
        - team_df (DataFrame): DataFrame containing season-level data 
            for the team provided from the 2020-21 regular season
    """

    per_mode = 'Totals'
    season = '2020-21'

    team_data = team.Team(headers=HEADERS,
                          endpoint='teamdashboardbygeneralsplits',
                          league_id=league_id,
                          season=season,
                          measure_type=measure_type,
                          per_mode=per_mode,
                          team_id=team_id)
    team_df = pd.DataFrame(team_data.data['OverallTeamDashboard'])

    return team_df
```

```python
TEAM_TO_ID_NBA = {'TOR': '1610612761', 'DEN': '1610612743',
                  'DET': '1610612765', 'NOP': '1610612740',
                  'MIL': '1610612749', 'GSW': '1610612744',
                  'SAS': '1610612759', 'POR': '1610612757',
                  'LAC': '1610612746', 'DAL': '1610612742',
                  'MEM': '1610612763', 'PHI': '1610612755',
                  'BOS': '1610612738', 'MIN': '1610612750',
                  'CHA': '1610612766', 'IND': '1610612754',
                  'ORL': '1610612753', 'MIA': '1610612748',
                  'HOU': '1610612745', 'SAC': '1610612758',
                  'UTA': '1610612762', 'BKN': '1610612751',
                  'ATL': '1610612737', 'PHX': '1610612756',
                  'WAS': '1610612764', 'NYK': '1610612752',
                  'OKC': '1610612760', 'LAL': '1610612747',
                  'CLE': '1610612739', 'CHI': '1610612741'}
```

```python
# take Los Angeles Lakers as example
team_df = construct_season_logs(TEAM_TO_ID_NBA['LAL'], 'Base')
team_df.iloc[0]
```

```python
oppo_df = construct_season_logs(TEAM_TO_ID_NBA['LAL'], 'Opponent')
```

```python
def pace_possession_calculator(game_df, prefix=''):
    """ This function calculates the possessions and pace
    from the DataFrame provided

    @param game_df (DataFrame): Single-row DataFrame corresponding to a team's season
    @param prefix (str): One of '' or 'OPP_'. If a team's
        data is provided, use ''. If an opponent's data
        is provided, use 'OPP_'

    Returns:
        - poss (float): Number of possessions in a season
        - pace (float): Number of possessions per 48 minutes
    """

    poss = sum(game_df[prefix + 'FGA']) \
        + (0.44 * sum(game_df[prefix + 'FTA'])) \
        - sum(game_df[prefix + 'OREB']) + sum(game_df[prefix + 'TOV'])

    pace = (48 * poss) / sum(game_df['MIN'])
    return poss, pace
```

```python
poss, pace = pace_possession_calculator(team_df, '')
o_rtg = 100 * sum(team_df['PTS']) / poss
print('Offensive Rating')
print(round(o_rtg, 2))
print('****************')
print('Pace')
print(round(pace, 2))
```

```python
d_poss, d_pace = pace_possession_calculator(oppo_df, 'OPP_')
d_rtg = 100 * sum(oppo_df['OPP_PTS']) / d_poss
print('Defensive Rating')
print(round(d_rtg, 2))
```

```python
def league_ratings_and_pace():
    """ This function loops all 30 teams and calculates offensive and defensive rating, and the pace

    Returns:
        - data_dict (dict): Dictionary storing
            ratings and pace data
    """

    data_dict = {}

    for team in TEAM_TO_ID_NBA:
        data_dict[team] = {}
        print('Pulling data for ' + team)
        team_df = construct_season_logs(TEAM_TO_ID_NBA[team], 'Base')
        time.sleep(2)
        oppo_df = construct_season_logs(TEAM_TO_ID_NBA[team], 'Opponent')

        # Calculating offensive side
        poss, pace = pace_possession_calculator(team_df, '')
        o_rtg = 100 * sum(team_df['PTS']) / poss

        # Calculating defensive side
        d_poss, d_pace = pace_possession_calculator(oppo_df, 'OPP_')
        d_rtg = 100 * sum(oppo_df['OPP_PTS']) / d_poss

        # Storing data
        data_dict[team]['Offensive Rating'] = o_rtg
        data_dict[team]['Defensive Rating'] = d_rtg
        data_dict[team]['Team Pace'] = pace
        data_dict[team]['Opponent Pace'] = d_pace
        time.sleep(2)

    return data_dict
```

```python
ratings_dict = league_ratings_and_pace()
```

```python
ratings_df = pd.DataFrame(ratings_dict).T.reset_index()
ratings_df.columns = ['Team'] + list(ratings_df)[1:]
ratings_df
```

## 3. Visualization

```python
p = ggplot(data=ratings_df) \
    + geom_point(mapping=aes(x='Offensive Rating',
                             y='Defensive Rating'), alpha=0.1) \
    + geom_text(mapping=aes(x='Offensive Rating',
                            y='Defensive Rating',
                            label='Team'),
                position=position_jitter(width=0.2,height=0.2), size=8) \
    + xlim(100, 120) \
    + ylim(100, 120)
p
```

* The best teams are located in the lower-right portion of the graph, with higher offensive ratings and lower defensive ratings (Utah, Philadelphia, Phoenix, Los Angeles Clippers, and Milwaukee).

```python
ratings_df = ratings_df.sort_values('Team Pace')
ratings_df['Teams'] = pd.Categorical(ratings_df['Team'], categories=list(ratings_df['Team']), ordered=True)
p = ggplot(data=ratings_df) \
    + geom_bar(mapping=aes(x='Teams',
                           y='Team Pace'), position='dodge', stat='identity') \
    + coord_flip()
p
```

* Washington, Milwaukee, and Golden State played at the fastest pace last season.
